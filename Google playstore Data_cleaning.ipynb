```python
numeric_features=[feature for feature in df_copy.columns if df_copy[feature].dtype!='O']
categorical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype=='O']

#print columns
print("We have {} numerical features :{}".format(len(numeric_features),numeric_features))
print("We have {} categorical features :{}".format(len(categorical_features),categorical_features))
```

```python
## chat gpt 's version

numerical_cols = df_copy.select_dtypes(include=['int64', 'float64','int32']).columns
categorical_cols = df_copy.select_dtypes(include=['object']).columns
```

```python
# Print the separated columns
print("Numerical Columns:", numerical_cols.tolist())
print("Categorical Columns:", categorical_cols.tolist())
```

```python
## proportion of count data on categorical columns

for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('-------------------------')
```

```python


# Plot KDE for each numerical column
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.kdeplot(df_copy[col], shade=True, color='blue')
    plt.title(f'KDE Plot for {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

## from the above we can make the following observations 

Rating,Year are left skewed

Reviews,Size,Installs,Price are right skewed 

Therefore outliers are present

```python
# For categorical columns
category=['Type','Content Rating']
# Set up subplots
n_cols = len(category)
fig, axes = plt.subplots(n_cols, 1, figsize=(8, 4 * n_cols))

# Plot count plots for each categorical column
for i, col in enumerate(category):
    sns.countplot(x=df_copy[col], ax=axes[i], palette='Set2')
    axes[i].set_title(f'Count Plot for {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Adjust layout to avoid overlapping
plt.tight_layout()
plt.show()
```

```python
## which category has the Largest installations?

# Group by 'Category' and sum the 'Installs' for each category
category_installs = df.groupby('Category')['Installs'].sum()

# Sort the categories by total installations in descending order
category_installs_sorted = category_installs.sort_values(ascending=False)

# Get the category with the largest installations
top_category = category_installs_sorted.idxmax()  # Category with largest installs
top_installs = category_installs_sorted.max()     # Largest install count

# Display the result
print(f"The category with the largest total installations is '{top_category}' with {top_installs} installations.")
```
